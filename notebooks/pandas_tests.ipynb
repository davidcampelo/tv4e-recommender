```python
# read CSV using a relative path
import pandas as pd
import requests
URL='http://api_mysql.tv4e.pt/api/recommendations/ratings'
data= requests.get(URL)
df=pd.DataFrame(data.json())

df = df[df.video_watched_type != 'forced']

# XXX use a function to calculate implicit rating considering the video lead time
df['rating_implicit'] = (df['video_watch_time']/100) * 0.3
df['rating_explicit'] = (df['rating_value'])         * 0.7
# If the explicit rating was negative, the implicit will be negative
df['rating_implicit'][df.rating_explicit < 0] = df['rating_implicit'] * -1
# create a new column to put implicit or explicit rating rating_value
df['overall_rating_value'] = df['rating_implicit'] + df['rating_explicit']

# implicit rating is the watched time / explicit rating is the like-0-dislike
df['rating_implicit'] = df['video_watch_time']/100
df['rating_explicit'] = df['rating_value']

df.loc[pd.isnull(df['rating_value']) & (df['video_watch_time'] >= 10),'overall_rating_value'] = (df['video_watch_time']/100) * 0.5

df.loc[(df['video_watch_time'] < 10),'overall_rating_value'] = 0
df['overall_rating_value']=df['overall_rating_value'].fillna(0)

import datetime
NOW = datetime.datetime.now()
df['rating_date_diff'] = NOW - pd.to_datetime(df['rating_date_creation'])
df = df[(df.rating_date_diff.dt.days < 14)]

# df = df[df.video_watch_time < 10]
df
```
